```python
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
import cv2
```

```python
image = cv2.imread('Lane1.png')
image2 = cv2.imread('anna_defending.png')
```

```python
import numpy as np
```

```python
image_copy = np.copy(image)
image_blub = np.copy(image2)
image_copy2 = cv2.cvtColor(image_blub, cv2.COLOR_BGR2RGB)
```

```python
# plt.imshow(image_copy)
plt.imshow(image_copy2)
```

```python
plt.imshow(image_copy)
```

```python
red_threshold = 0
green_threshold = 0
blue_threshold = 0
rgb_threshold = [red_threshold, green_threshold, blue_threshold]
thresholds2 = (image2[:,:,0]< rgb_threshold[0]) | (image2[:,:,1]< rgb_threshold[1]) | (image2[:,:,2]< rgb_threshold[2])
thresholds = (image[:,:,0]< rgb_threshold[0]) | (image[:,:,1]< rgb_threshold[1]) | (image[:,:,2]< rgb_threshold[2])
```

```python
image_copy[thresholds] = [0,0,0]
image_copy2[thresholds2] = [0,0,0]
```

```python
plt.imshow(image_copy2)
```

```python
plt.imshow(image_copy)
```

```python
edges2 = cv2.Canny(image_copy2, 150, 200)
edges = cv2.Canny(image_copy, 150, 200)
```

```python
plt.imshow(edges)
```

```python
plt.imshow(edges2)
```

```python
cv2.imwrite( "quidditch_edges.png", edges2 )
cv2.imwrite( "lane_edges.png", edges )
```
